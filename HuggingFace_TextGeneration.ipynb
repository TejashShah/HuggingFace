```python
from transformers import pipeline, set_seed
```

```python
import textwrap  
import numpy as np 
import matplotlib.pyplot as plt
from pprint import pprint as pp
```

```python
!wget -nc https://raw.githubusercontent.com/lazyprogrammer/machine_learning_examples/master/hmm_class/robert_frost.txt
```

```python
!cat robert_frost.txt 
```

```python
lines = [ line.rstrip() for line in open('robert_frost.txt')] 
lines = [ line for line in lines if len(line) > 0]
```

```python
gen_pipeline = pipeline("text-generation") 

set_seed(1234) 
lines[0]
```

```python
gen_pipeline(lines[0])
```

```python
pp(_)
```

```python
pp(gen_pipeline(lines[0], max_length=20))
```

```python
pp(gen_pipeline(lines[0], num_return_sequences=3, max_length=20))
```

```python
def wrap(x):
  return textwrap.fill(x, replace_whitespace=False, fix_sentence_endings=True)
```

```python
out = gen_pipeline(lines[0], max_length=30)
print(wrap(out[0]['generated_text']))
```

```python
prev = 'Two roads diverged in a yellow wood, including one that blocked the' + \
  ' road leading to another intersection in the middle of the city.'

out = gen_pipeline(prev + '\n' + lines[2], max_length=60)
print(wrap(out[0]['generated_text']))
```

```python
prev = 'Two roads diverged in a yellow wood, including one that blocked the' + \
  ' road leading to another intersection in the middle of the city.\n' + \
  'And be one traveler, long I stood in front of the burning wreckage. ' + \
  'That\'s what I had written on Twitter a few minutes ago.'

out = gen_pipeline(prev + '\n' + lines[4], max_length=90)
print(wrap(out[0]['generated_text']))
```

```python
prompt = "Neural networks with attention have been used with great success"  + \
  " in natural language processing."
out = gen_pipeline(prompt, max_length=300)
print(wrap(out[0]['generated_text']))
```
